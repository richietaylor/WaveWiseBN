# Earthquake Example from class

```python
from pylab import *
import matplotlib.pyplot as plt
import os
import pyAgrum as gum
import pyAgrum.lib.notebook as gnb

bn=gum.BayesNet('EarthquakeEg')
print(bn)
```

```python
bn=gum.fastBN("b->a<-e;j<-a->m")
bn
```

```python
bn.cpt("e").fillWith([0.001,0.999])
bn.cpt("b").fillWith([0.002,0.998])

bn.cpt("a")[{'e': 1, 'b': 1}] = [0.95, 0.05]
bn.cpt("a")[{'e': 0, 'b': 1}] = [0.94, 0.06]
bn.cpt("a")[{'e': 1, 'b': 0}] = [0.29, 0.71]
bn.cpt("a")[{'e': 0, 'b': 0}] = [0.001, 0.999]

bn.cpt("j")[{"a":1}]= [0.9,0.1]
bn.cpt("j")[{"a":0}]= [0.05,0.95]
bn.cpt("m")[{"a":1}]= [0.7,0.3]
bn.cpt("m")[{"a":0}]= [0.01,0.99]
bn
```

```python
gnb.showInference(bn,evs={"b":0})
```
